## 2. ***Data Wrangling and Exploratory Data Analysis(EDA)***

```python
df = dataset.copy()
```

```python
# Only keeping the area codes in the area_code column --> 415,408,510
df['area_code'] = df['area_code'].str.extract(r'(\d+)')
```

```python
#Understanding the churn distribution
df['churn'].value_counts()
```

```python
df.groupby('area_code')['churn'].value_counts(normalize=True)
```

```python
df.groupby('state')['churn'] \
  .value_counts(normalize=True) \
  .loc[:, 'yes'] \
  .sort_values(ascending=False) \
  .head(10)
```

```python
df['churn_flag'] = df['churn'].map({'no': 0, 'yes': 1})
```

```python
df.groupby('number_customer_service_calls')['churn_flag'].mean()
```

**Churn probability remains stable (~10%) up to 3 service calls, but increases sharply beyond that, reaching over 60% for 5+ calls. This indicates unresolved customer issues are a major driver of churn.**

Let's analyze the behavioral patterns of the customers

```python
#Avg duration of calls for churned customers
df['avg_day_call_duration'] = df['total_day_minutes'] / df['total_day_calls']
df['avg_eve_call_duration'] = df['total_eve_minutes'] / df['total_eve_calls']
df['avg_night_call_duration'] = df['total_night_minutes'] / df['total_night_calls']
df['avg_intl_call_duration'] = df['total_intl_minutes'] / df['total_intl_calls']
```

```python
df['total_calls'] = df['total_day_calls'] + df['total_eve_calls'] + df['total_night_calls'] + df['total_intl_calls']
```

```python
df['avg_call_duration'] = df['total_usage']/df['total_calls']
```

```python
df['day_usage_share'] = df['total_day_minutes'] / df['total_usage']
df['eve_usage_share'] = df['total_eve_minutes'] / df['total_usage']
df['night_usage_share'] = df['total_night_minutes'] / df['total_usage']
df['intl_usage_share'] = df['total_intl_minutes'] / df['total_usage']
```

###***Key EDA Insights***


1. **Customer Service Calls emerged as the strongest churn signal**

Customers with a higher number of customer service calls, especially those with more than 3 calls, showed a significantly higher churn tendency. This indicates that unresolved issues, dissatisfaction, or repeated support dependency are strongly associated with customer attrition.

2. **International Plan customers showed elevated churn risk**

Customers subscribed to the International Plan had a noticeably higher churn rate compared to those without it. This suggests that international-plan users may be more price-sensitive, experience service-related issues, or have expectations that are not being met.

3. **The combination of International Plan + High Service Calls was especially risky**

One of the strongest behavioral churn patterns observed was among customers who had:

International Plan = Yes
Customer Service Calls > 3

This group showed a particularly high churn rate, indicating a compounding effect of product-related friction and service dissatisfaction.

4. **Low Usage customers with repeated support calls also had high churn**

Customers with lower overall telecom usage but high customer service interactions were also more likely to churn. This may indicate disengaged or low-value customers who encounter friction early and do not develop strong service dependency.

5. **Total Usage and Total Charges were useful supporting churn indicators**

Customers differed meaningfully across usage and billing-related behavior, making variables such as:

Total Usage
Total Charges

important supporting features for churn understanding. These variables likely capture customer engagement intensity and account value.
